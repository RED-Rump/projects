## Load Pre-trained BERT Model

### Subtask:
Load a pre-trained BERT model and its corresponding tokenizer from the 'transformers' library. This model will be used to generate embeddings for the text documents.


**Reasoning**:
The subtask requires loading a pre-trained BERT model and its tokenizer. I will import the necessary classes from the `transformers` library and then load the 'bert-base-uncased' tokenizer and model into the specified variables.



```python
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
model = AutoModel.from_pretrained('bert-base-uncased')

print("BERT tokenizer and model loaded successfully.")
```

**Reasoning**:
The previous step successfully loaded the BERT tokenizer and model. The next logical step in a BERT-based topic modeling pipeline is to use these components to generate embeddings for the preprocessed text documents. This will convert the text into numerical representations suitable for clustering.



```python
import torch

def get_bert_embeddings(texts, tokenizer, model):
    encoded_input = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        model_output = model(**encoded_input)
    # Use the embeddings of the [CLS] token as the sentence embedding
    return model_output.last_hidden_state[:, 0, :].numpy()

# Join the preprocessed tokens back into strings for BERT input
clean_documents = [" ".join(doc) for doc in process_docs]
bert_embeddings = get_bert_embeddings(clean_documents, tokenizer, model)
print("BERT embeddings generated successfully.")
```

**Reasoning**:
The previous code failed because the `process_docs` variable was not defined in the current execution environment, leading to a `NameError`. To fix this, I will re-include the definitions for `documents` and the preprocessing logic, which creates `process_docs`, within this code block. This ensures all necessary dependencies are met before generating BERT embeddings.



```python
import torch
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Re-define documents and preprocess_text to ensure process_docs is available
documents = [
    "The quick brown fox jumps over the lazy dog.",
    "A brown bear is often seen near the river.",
    "The dog barks loudly at the cat.",
    "Cats love to chase mice and play with string.",
    "A fox and a dog are both canids."
]

nltk.download('stopwords', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    return filtered_tokens

process_docs = [preprocess_text(doc) for doc in documents]

def get_bert_embeddings(texts, tokenizer, model):
    encoded_input = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        model_output = model(**encoded_input)
    return model_output.last_hidden_state[:, 0, :].numpy()

clean_documents = [" ".join(doc) for doc in process_docs]
bert_embeddings = get_bert_embeddings(clean_documents, tokenizer, model)
print("BERT embeddings generated successfully.")
```

## Cluster Embeddings to Form Topics

### Subtask:
Apply a clustering algorithm, such as K-Means or HDBSCAN, to the generated BERT embeddings. Each cluster will represent a distinct topic.


**Reasoning**:
The subtask requires applying a clustering algorithm to the generated BERT embeddings. I will use KMeans as instructed, initializing it with the specified parameters, fitting it to `bert_embeddings`, and storing the cluster labels.



```python
from sklearn.cluster import KMeans

kmeans_model = KMeans(n_clusters=3, random_state=42, n_init=10) # Added n_init to suppress warning
cluster_labels = kmeans_model.fit_predict(bert_embeddings)

print("BERT embeddings clustered successfully.")
print("Cluster Labels:", cluster_labels)
```

**Reasoning**:
The subtask of clustering the BERT embeddings is complete. The next logical step is to associate the original documents with their assigned cluster labels to begin interpreting the identified topics.



```python
clustered_documents = pd.DataFrame({'Document': documents, 'Cluster': cluster_labels})
print("Documents with assigned cluster labels:")
print(clustered_documents)
```

**Reasoning**:
The previous code failed because the pandas library was not imported. To fix this, I need to add 'import pandas as pd' before creating the DataFrame.



```python
import pandas as pd

clustered_documents = pd.DataFrame({'Document': documents, 'Cluster': cluster_labels})
print("Documents with assigned cluster labels:")
print(clustered_documents)
```

**Reasoning**:
Now that BERT embeddings have been clustered, the next logical step is to interpret these clusters as topics. I will extract the most frequent words from the documents belonging to each cluster to represent and display the topics found by the BERT-based clustering.



```python
from collections import Counter

# Combine documents and their assigned clusters
clustered_data = list(zip(process_docs, cluster_labels))

NUM_TOPICS = len(set(cluster_labels))
topics_bert = []

for i in range(NUM_TOPICS):
    # Get all words from documents in the current cluster
    cluster_words = [word for doc_tokens, cluster_id in clustered_data if cluster_id == i for word in doc_tokens]

    # Count word frequencies
    word_counts = Counter(cluster_words)

    # Get the top 10 most common words for the topic
    top_words = [word for word, count in word_counts.most_common(10)]
    topics_bert.append(top_words)
    print(f"BERT Topic {i}: {', '.join(top_words)}")

print("BERT-based topics extracted successfully.")
```
